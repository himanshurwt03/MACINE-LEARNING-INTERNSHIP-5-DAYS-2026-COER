```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="age", y="charges")
plt.title("age vs charges")
plt.xlabel("age")
plt.ylabel("charges")
plt.show()
```
