```python
df.fillna({'Age':18,'Gender':'Male'},inplace=True)
df.isna().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Age'].isnull().sum()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode())
df['Gender'].isnull().sum()
```
